```python
from tokenizer import decoder
from generate import choseToken, vocab

for i in range(5):
  tokenActual = "256"
  while True:
    pair = choseToken(tokenActual)
    if(pair == None):
      break
    pair = pair.replace("(", "").replace(")", "").split(",")
    tokenOld = pair[0]
    tokenActual = pair[1].replace(" ", "")
    # print(tokenActual)
    print(decoder([int(tokenOld)], vocab), end="")
  print("")
  
```
