Agora, com o DataFrame que foi construído, os modelos irão se basear neles para realizar suas predições.

A partir dos modelos treinados, o documento de testes será rodado para ser possível a realização das análises sobre cada um dos modelos.

A primeira análise será com o modelo Multinomial Naive Bayes:

```python
import find_best_hyperparameters as fbh
import sklearn as sk

f = open("test.csv", "r")
temp = p.read_csv(f)
testX, testY = temp.iloc[:100, :-2], temp.iloc[:100, -1]
f.close()

# Get all frases with the predicted class 
predicted = []
for text in testX.loc[:, "text"]:
  # Get the class prediction 
  predicted.append(int(fbh.Naive_Bayes_Classifier(text)))

# Get the accuracy
print(sk.metrics.accuracy_score(testY, predicted))
```

O resultado que foi obtido consegue demonstrar que o modelo que foi construído foi capaz de obter os resultados esperados com uma precisão de 49 %, o que ainda não é muito conclusivo sobre o modelo, mas já da uma certa base sobre o seu funcionamento.

```python
print(sk.metrics.f1_score(testY, predicted, average="macro"))
print(sk.metrics.f1_score(testY, predicted, average="micro"))
```

Nessa outra etapa foi obtido o F1 Score do modelo, que reune o cálculo da acurácia que foi vista no cálculo anterior com o conceito de recall, que determina se o modelo está sendo capaz de detectar os exemplos positivos. Com o resultado obtido, é possível ver que o modelo se encontra bem equilibrado, tanto no aspecto macro quanto no micro.
